```python
# ttest_ind is a two-sided test for the null hypothesis that 2 independent samples
# have identical average (expected) values


from numpy import *
from scipy import stats
import pylab
import matplotlib.pyplot as plt
%matplotlib inline
drugA=recfromcsv('C:\\Users\\19293\\Desktop\\Data Science\\W5L2\\data2\\drug-A.csv')
placeboA=recfromcsv('C:\\Users\\19293\\Desktop\\Data Science\\W5L2\\data2\\placebo-A.csv')
drugA1=array(drugA, dtype=float)
placeboA1=array(placeboA, dtype=float)
t_statistic,p_value=stats.ttest_ind(drugA1,placeboA1)
print('The p_value is',p_value)
print('The t-statistic value is',t_statistic)
```

```python
#test

n,bins,temp=plt.hist(drugA1,5,color='blue', edgecolor='k',normed=1)
xx=linspace(0,2,100)
sigma=std(drugA1)
mu=mean(drugA1)
y = sqrt(1/(2*pi*sigma**2))*exp(-(xx-mu)**2/(2*sigma**2))
plt.plot(xx, y,'k--')
pylab.setp(temp, linestyle='-', linewidth=1)
plt.xlabel('drugA1')
plt.ylabel('frequency')
plt.xticks(bins)
plt.title('drugA1 histrogram')
plt.show()
```

```python
#two-tailed normal test
#trilo1
k2,p=stats.normaltest(drugA1)
print('skew=',k2)
print('H0-drugA comes from a normal distrubution:')
if p<0.05:
    print('Reject in favor of H1')
else:
    print('accept')

plt.hist(drugA1,bins=5)
plt.legend(('p-value=%.2g'%p,))
plt.show()
```

```python
# DAgostinos K^2 Test
# D'Agostino and Pearson's Test on drug A
from numpy.random import seed
from numpy.random import randn
from scipy.stats import normaltest
# seed the random number generator
seed(1)
# generate univariate observations
# normality test
stat, p = normaltest(drugA1)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
    
```
