```python
from PIL import Image as pili, ImageDraw as pild, ImageFont as pilf,  ImageOps as piliops

#import the library opencv
import cv2
#globbing utility.
import glob
#select the path
path = "C:/Users/19293/Desktop/comic/images/"
count=0
for filename in os.listdir(path):
    print (filename)
    TINT_COLOR = (0, 0, 0)  # Black
    TRANSPARENCY = .25  # Degree of transparency, 0-100%
    OPACITY = int(255 * TRANSPARENCY)
    img = pili.open("C:/Users/19293/Desktop/comic/images/" + filename).convert('RGBA')
    caption = data.values[count][0]
    print (caption)
    overlay = pili.new('RGBA', img.size, TINT_COLOR+(0,))
    draw = pild.Draw(overlay)

    font = pilf.truetype("NotoSansDevanagari-Bold.ttf", 60)
    text = caption
    w, h = font.getsize(text)
    num_lines = len(text.split('\n'))
    print(num_lines, w, h)
    x, y = 30, img.height - (num_lines-0.1*num_lines)*h
    draw.rectangle((x, y, x + img.width - 200, y + (num_lines-0.1*num_lines)*h), fill=TINT_COLOR+(OPACITY,))
    draw.text((x, y), text, fill=(209, 239, 8), font=font, align = 'center')

    
    # Alpha composite these two images together to obtain the desired result.
    img = pili.alpha_composite(img, overlay)
    img = img.convert("RGB") # Remove alpha for saving in jpg format.

    #saving images in new folder
    piliops.expand(img, border=(30, 20), fill='red').save("C:/Users/19293/Desktop/comic/images2/" + filename)
    count = count + 1
```

```python
from os import listdir
from os.path import isfile, join
from PIL import Image
import fitz
import numpy as np


mypath = "C:/Users/19293/Desktop/comic/images2/"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

doc = fitz.open()    #create new pdf

cur_num_cols = np.random.randint(2, 4)
curY=0
x=0
row = 1
while x<len(onlyfiles):
    if x % 15 == 0:
        curY = 0
        print("new page!")
        doc.insertPage(-1)  # insert a new page in front of page n
        row = 1
    page = doc[-1]
    pgWidth = page.MediaBox.width
    pgHeight = page.MediaBox.height
    curX=0
    #print("here")
    for j in range(0, cur_num_cols):
        #print("j=",j)
        #print("curX=",curX)
        #print("curY=",curY)
        #print(curX + curY + int(pgHeight/5))
        img = onlyfiles[x]
        rect = fitz.Rect(curX, curY, curX + int (pgWidth/cur_num_cols), curY + int(pgHeight/5)) # where we want to put the image
        curX+=int(pgWidth/cur_num_cols)
        pix = fitz.Pixmap(img)        # any supported image file
        page.insertImage(rect, pixmap=pix, overlay=True, keep_proportion=False)   # insert image
        x=x+1
    row+=1
    #print(x)
    curY+=int(pgHeight/5)
    if row==5:
        #print("x=", x)
        cur_num_cols = 15 - (x % 15)
        #print("cur_num_cols", cur_num_cols)        
    else:
        cur_num_cols = np.random.randint(2, 4)
    
    
doc.save("mybook.pdf")      
  
```
